```python
# List
List = ["every day", "last week", "last year", "last month", "right now", "yesterday", "today", "yesterday", "now", "this year", "in 1990", "while", "I think Marie will arrive in Stuttgart at 10 am.", "I am going to go to school.", "I buy a new car.", "Wait, I will help you.", "It is raining right now.","We are going to play football at the birthday party.", "The will-future is about future plans. The going-to-future is always about spontaneous decisions and actions you can not control.","The will-future is about spontaneous decisions and actions you can not control. The going-to-future is always about future plans.","You can use both whenever you want.","When actions are common or happen repeatedly.","When the action finished in the past.", "When you plan an action for the future.","When something might happen in the future.","When you just finished with something.", "When something is happening right now.","When you started something in the past and did not finish yet.","When the action finished in the past.","When the situation you are talking about happens right now.","When you plan something for the future.","When the action happened in the middle of another action.", "When the action finished 10 years ago.", "in 1990", "while", "I think Marie will arrive in Stuttgart at 10 am.", "I will go to school.", "I will buy a new car.", "Wait, I am going to help you.", "It is going to rain tomorrow.", "We are going to play football at the birthday party.", "The will-future is about future plans. The going-to-future is always about spontaneous decisions and actions you can not control.", "The will-future is about spontaneous decisions and actions you can not control. The going-to-future is always about future plans.", "You can use both whenever you want.", "When actions are common or happen repeatedly.", "When the action finished in the past.", "When you plan an action for the future.","When something might happen in the future.", "When you just finished with something.","When something is happening right now.", "When you started something in the past and did not finish yet.","When the action finished in the past.", "When the situation you are talking about happens right now.", "When you plan something for the future.", "When the action happened in the middle of another action.","When the action finished 10 years ago."]
questions = ["What is a signal word for the simple present?",
             "What is a signal word for the present progressive?",
             "What is a signal word for the simple past?",
             "What is a signal word for the past progressive?",
             "Which of these sentences is in the will-future?",
             "Which of these sentences is in the going-to-future?",
             "What is the difference between the will and going-to future tenses?",
             "When do you use the simple present?",
             "When do you use the present progressive?",
             "When do you use the simple past?"
             ]
options = ["a", "b", "c"]
correct = ["a", "b", "b", "c", "a", "c", "b", "a", "c", "b"]


class Score():
    s = 0

def scoreup():
    Score.s = Score.s + 1


def scoredown():
    Score.s = Score.s - 1


def resetscore():
    Score.s = 0


# Questions
while True:
    i = 0
    for q in questions:
        print(q)
        for ans in range(3):
            print(str(options[ans]) + " " + str(List[i]))
            i += 1
        answer = input("What is your answer?" + " ")
        if answer == correct[questions.index(q)]:
            print("That is correct!")
            scoreup()
            print("\33[1mYour score is" + " " + str(Score.s) + "\33[m.")
        else:
            print("Sorry, that is wrong. ")
            scoredown()
            print("\33[1mYour score is" + " " + str(Score.s) + "\33[m.")
            if Score.s <= -2:
                print("You have to start again!")
                break
    if Score.s == -2:
        resetscore()
    else:
        print("\33[1mGood Gob! You are finished! Your final score is " + str(Score.s) + "\33[m.")
        break
```
